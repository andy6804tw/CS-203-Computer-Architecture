```python
# import matplotlib相關套件
import matplotlib.pyplot as plt



# 使用月份當做X軸資料

month = [1,2,3,4]

# 使用台G電的某年每月收盤價當第一條線的資料

cache512 = [0.857885,0.534479,0.374775,0.29920]

# 使用紅海的某年每月收盤價當第二條線的資料

cache1024 = [0.857086,0.533880,0.374175,0.298021]

 

# 設定圖片大小為長15、寬10

plt.figure(figsize=(10,5),dpi=100,linewidth = 2)

# 把資料放進來並指定對應的X軸、Y軸的資料，用方形做標記(s-)，並指定線條顏色為紅色，使用label標記線條含意

plt.plot(month,cache512,'s-',color = 'r', label="TSMC")

# 把資料放進來並指定對應的X軸、Y軸的資料 用圓形做標記(o-)，並指定線條顏色為綠色、使用label標記線條含意

plt.plot(month,cache1024,'o-',color = 'g', label="FOXCONN")

 

# 設定圖片標題，以及指定字型設定，x代表與圖案最左側的距離，y代表與圖片的距離

plt.title("Python範例", x=0.5, y=1.03)

# 设置刻度字体大小

# plt.xticks(fontsize=20)

# plt.yticks(fontsize=20)

# 標示x軸(labelpad代表與圖片的距離)

plt.xlabel("month", labelpad = 15)

# 標示y軸(labelpad代表與圖片的距離)

plt.ylabel("price", labelpad = 20)

# 顯示出線條標記位置

plt.legend(loc = "best")

# 畫出圖片

plt.show()
```
